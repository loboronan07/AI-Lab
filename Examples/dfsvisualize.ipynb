```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os
```

```python
os.mkdir('dfs')
```

```python
def display_gif(fn):
    from IPython import display
    return display.HTML('<img src="{}">'.format(fn))
```

```python
class Graph:
    def __init__(self, n):
        self.n = n
        self.adjmat = []
        
        for i in range(n):
            self.adjmat.append([0 for j in range(n)])
        
        self.g = nx.Graph()
    
    def addedge(self, u, v):
        self.adjmat[u][v] = 1
        self.adjmat[v][u] = 1
        self.g.add_edge(u, v, weight=1)
        
    def visualize(self):
        pos = nx.kamada_kawai_layout(self.g)
        nx.draw(self.g, pos, with_labels=False, node_size=300, width=1)
        plt.show()
        
    def hightlight(self, traversal):
        pos = nx.kamada_kawai_layout(self.g)
        for i in range(len(traversal)):
            edges = []
            for j in range(len(traversal[i])-1):
                edges.append((traversal[i][j], traversal[i][j+1]))
            
            nx.draw_networkx(self.g, pos, with_labels=False, node_size=500, width=3)
            nx.draw_networkx_nodes(self.g, pos, traversal[i], node_size=500, node_color='r')
            nx.draw_networkx_edges(self.g, pos, edgelist=edges, edge_color='g', width=3)
            plt.show()        
```

```python
def dfs_start(graph, start):
    traversal = []
    visitedhist = []
    stack = []
    
    visited = set()
    visited.add(start)
    
    dfs(graph, start, stack, traversal, visited, visitedhist)
    
    return traversal, visitedhist


def dfs(graph, curr, stack, traversal, visited, visitedhist):
    stack.append(curr)
    traversal.append(stack[:])
    visitedhist.append(list(visited))
    
    for i in range(graph.n):
        if graph.adjmat[curr][i] == 1:
            if i not in visited :
                visited.add(i)
                dfs(graph, i, stack, traversal, visited, visitedhist)
                traversal.append(stack[:])
                visitedhist.append(list(visited))
    
    stack.pop()
```

```python
def creategif(graph, start, name):
    traversal, visited = dfs_start(graph, start)
    pos = nx.kamada_kawai_layout(graph.g)
    
    frames = []
    
    for i in range(len(traversal)):
        edges = []
        for j in range(len(traversal[i])-1):
            edges.append((traversal[i][j], traversal[i][j+1]))
        
        plt.figure(figsize=(20, 15))
        nx.draw_networkx(graph.g, pos, with_labels=False, node_size=1000, width=6)
        nx.draw_networkx_nodes(graph.g, pos, visited[i], node_size=1000, node_color='b')
        nx.draw_networkx_nodes(graph.g, pos, traversal[i], node_size=1000, node_color='r')
        nx.draw_networkx_edges(graph.g, pos, edgelist=edges, edge_color='g', width=6)
        plt.savefig(f'./dfs/img_{i}.png', 
                transparent = False,  
                facecolor = 'white'
               )
        plt.close()
        
        image = imageio.v2.imread(f'./dfs/img_{i}.png')
        frames.append(image)
    
    imageio.mimsave(f'./dfsgiphy{name}.gif',
                frames,
                fps = 5)
```

```python
g = Graph(81)

for u in range(72):
    g.addedge(u, u+9)
    
u = 0
while(u < 80):
    if((u+1) % 9 != 0):
        g.addedge(u, u+1)
    u += 1

g.visualize()
```

```python
creategif(g, 40, 'grid81')
display_gif('dfsgiphygrid81.gif')
```
